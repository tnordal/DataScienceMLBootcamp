## Matplotlib Essentials Part 1

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(11)
y = x ** 2
```

```python
plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Figure x and x ** 2')
# plt.show()
```

```python
plt.subplot(1, 3, 1)
plt.plot(x, y, 'r+-')

plt.subplot(1, 3, 2)
plt.plot(y, x, 'g--')

plt.subplot(1, 3, 3)
plt.plot(x, y, 'r*-')
plt.plot(y, x, 'g*-')

plt.tight_layout()
```
