## Matplotlib Essentials Part 1

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(11)
y = x ** 2
```

```python
plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Figure x and x ** 2')
# plt.show()
```

```python
plt.subplot(1, 3, 1)
plt.plot(x, y, 'r+-')

plt.subplot(1, 3, 2)
plt.plot(y, x, 'g--')

plt.subplot(1, 3, 3)
plt.plot(x, y, 'r*-')
plt.plot(y, x, 'g*-')

plt.tight_layout()
```

## Matplotlib Essentials Part 2 Object Oriented

```python
fig1 = plt.figure()
axes = fig1.add_axes([0.1, 0.1, 0.9, 0.9])
axes.plot(x, y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Figure')
```

```python
fig2 = plt.figure()
axes_main = fig2.add_axes([0.1, 0.1, 0.9, 0.9])
axes_inset = fig2.add_axes([0.3, 0.5, 0.3, 0.4])
axes_main.plot(x, y, 'b')
axes_inset.plot(y, x, 'r')

axes_main.set_xlabel('Main X')
axes_main.set_ylabel('Main Y')
axes_main.set_title('Main Title')

axes_inset.set_xlabel('Inset X')
axes_inset.set_ylabel('Inset Y')
axes_inset.set_title('Inset Title')
```

```python
fig, axes = plt.subplots()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True)
plt.tight_layout()
```

```python
fig1, axes1 = plt.subplots(nrows=2, ncols=1)
fig2, axes2 = plt.subplots(nrows=1, ncols=2)
```

```python
type(axes1), type(axes2)
```

```python
for ax in axes1:
    ax.plot(x, y)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Title')
fig1.tight_layout()
fig1
```

```python
axes2[0].plot(x, y, 'r')
axes2[1].plot(y, x, 'g')
fig2.tight_layout()
fig2
```

## Matplotlib Essentials Part 3 Figure Size, Aspect Ratio and DPI

```python
fig = plt.figure(figsize=(10, 5), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(10, 5))
axes.plot(x, y, 'r')
```

```python
fig.savefig('figname.png', dpi=100)
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x**2, label='x**2')
axes.plot(x, x**3, label='x**3')
axes.legend(loc=0)
```

```python
fig , axes = plt.subplots()
axes.plot(x, x+1, 'r.-')
axes.plot(x, x+2, 'g--')
```

```python
fig , axes = plt.subplots()
axes.plot(x, x+1, color='blue', alpha=0.7)
axes.plot(x, x+2, color='#8B008B')
axes.plot(x, x+3, color='#FF8C00')
```

```python
fig , axes = plt.subplots(figsize=(10, 8))
axes.plot(x, y, color='purple', lw=4, ls='-', marker='s', markersize=18, markerfacecolor='yellow', markeredgewidth=4,
         markeredgecolor='red')
```

```python
fig , axes = plt.subplots(nrows=1, ncols=3, figsize=(10, 3))
axes[0].plot(x, y)
axes[0].set_title('Default')

axes[1].plot(x, y)
axes[1].axis('tight')
axes[1].set_title('Tight')

axes[2].plot(x, y)
axes[2].set_title('Custom')
axes[2].set_xlim([1, 4])
axes[2].set_ylim([0, 50])
```
